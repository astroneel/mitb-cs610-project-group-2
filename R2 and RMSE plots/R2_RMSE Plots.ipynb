```python
import pandas as pd
import matplotlib.pyplot as py
import numpy as np
```

```python
with_covid_R2 = pd.read_csv('data/R2 data.csv').iloc[:-1 , :]
with_covid_RMSE = pd.read_csv('data/RMSE data.csv').iloc[:-1 , :]
without_covid_R2 = pd.read_csv('data/R2 data (before 2019).csv').iloc[:-1 , :]
without_covid_RMSE = pd.read_csv('data/RMSE data (before 2019).csv').iloc[:-1 , :]
```

```python
_, ax1 = py.subplots()
ax1.set_title("R2 Plot")
ax1.plot(with_covid_R2['ANN'], label = 'ANN')
ax1.plot(with_covid_R2['Gradient Boosting'], label = 'Gradient Boosting')
ax1.plot(with_covid_R2['Random Forrest'], label = 'Random Forrest')
ax1.plot(with_covid_R2['Lasso Regression'], label = 'Lasso Regression')
ax1.plot(with_covid_R2['Linear Regression'], label = 'Linear Regression')
ax1.legend()
```

```python
_, ax2 = py.subplots()
ax2.set_title("RMSE Plot")
ax2.plot(with_covid_RMSE['ANN'], label = 'ANN')
ax2.plot(with_covid_RMSE['Gradient Boosting'], label = 'Gradient Boosting')
ax2.plot(with_covid_RMSE['Random Forrest'], label = 'Random Forrest')
ax2.plot(with_covid_RMSE['Lasso Regression'], label = 'Lasso Regression')
ax2.plot(with_covid_RMSE['Linear Regression'], label = 'Linear Regression')
ax2.legend()
```

```python
_, ax3 = py.subplots()
ax3.plot(without_covid_R2['ANN'], label = 'before Covid')
ax3.plot(with_covid_R2['ANN'], label = 'with Covid')
ax3.legend()
```

```python
_, ax4 = py.subplots()
ax4.plot(without_covid_R2['Random Forrest'], label = 'before Covid')
ax4.plot(with_covid_R2['Random Forrest'], label = 'with Covid')
ax4.legend()
```

```python
_, ax5 = py.subplots()
ax5.plot(without_covid_R2['Gradient Boosting'], label = 'before Covid')
ax5.plot(with_covid_R2['Gradient Boosting'], label = 'with Covid')
ax5.set_ylim(bottom=None, top=1.0)
ax5.legend()
```
